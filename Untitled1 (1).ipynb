```python
import pandas as pd
import numpy as np
```

```python
credits=pd.read_csv("credits.csv")
```

```python
movies=pd.read_csv("movies.csv")
```

```python
credits2=credits.rename(columns={"movie_id":"id"})
```

```python
df=movies.merge(credits2,on="id")
```

```python
columns=["id","original_title","overview"]
```

```python
df2=df[columns]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf=TfidfVectorizer(min_df=4,max_features=None,strip_accents="unicode",analyzer="word",token_pattern="r\w{1,}",ngram_range=(1,3),stop_words="english")
```

```python
df2["overview"]=df2["overview"].fillna(" ")
```

```python
tv=tfidf.fit_transform(df2["overview"])
```

```python
tv
```

```python
tv.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
```

```python
sigmoid=sigmoid_kernel(tv,tv)
```

```python
sigmoid[0]
```

```python
arjun=pd.Series(df2.index,index=df2["original_title"]).drop_duplicates()
```

```python
sorted(list(enumerate(sigmoid[arjun["Titanic"]])),key=lambda x:x[1],reverse=True)
```

```python
def make_prediction(original_title,sigmoid=sigmoid):
    index=arjun[original_title]
    sigma_values=list(enumerate(sigmoid[index]))
    sigma_values=sorted(sigma_values,key=lambda x:x[1],reverse=True)
    sigma_values=sigma_values[1:11]
    arjuna=[i[0] for i in sigma_values]
    return df2["original_title"].iloc[arjuna] 
    
```

```python
make_prediction("Titanic")
```
